```python
sns.displot(tips['total_bill'],kde=False)
```

```python
#jointplot allows you to compare with what kind of parameters of bivariate data we want.
#scatter
#reg
#resid
#kde
#hex
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#pairplot will plot pairwise relationship across an entire dataframe(for the numerical column) and supports a color hue argument(for the categorical column).
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='brg')
```

```python
#Rugplot just draw small lines denisty wise:
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
# Matrix form for correction data
tips.corr(numeric_only=True)
```

```python
sns.heatmap(tips.corr(numeric_only=True),cmap='magma',annot=True)
```

```python
sns.heatmap(tips.corr(numeric_only=True),annot=True)
```

```python
#It uses hierachal clustering to produce a clustered version of the Heatmap.
sns.clustermap(tips.corr(numeric_only=True),annot=True)
```
